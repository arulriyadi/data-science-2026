# Praktikum Pertemuan 4: Statistika Dasar & Analisis Data


## Langkah 1: Load & Inspect Dataset

Muat dataset **Iris** dari Seaborn, lalu tampilkan `shape`, `dtypes`, `head()`, dan `describe()` untuk memahami struktur data sebelum dianalisis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = sns.load_dataset('iris')
print('Shape:', df.shape)
print(df.dtypes)
print(df.head())
print(df.describe().round(3))
```

## Langkah 2: Statistik Deskriptif Lengkap

Hitung **mean, median, std, skewness, dan kurtosis** untuk setiap kolom numerik.


```python
for col_name in df.select_dtypes(include='number').columns:
    col = df[col_name]
    print(f'\n=== {col_name} ===')
    print(f'  Mean     : {col.mean():.3f}')
    print(f'  Median   : {col.median():.3f}')
    print(f'  Std Dev  : {col.std():.3f}')
    print(f'  Skewness : {col.skew():.3f}')
    print(f'  Kurtosis : {col.kurt():.3f}')
```

## Langkah 3: Analisis Distribusi (Histogram + KDE)

Histogram + KDE pada `sepal_length` dengan garis **mean** dan **median**.


```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df['sepal_length'], kde=True, color='#028090', bins=20, ax=ax)
ax.axvline(df['sepal_length'].mean(), color='red', linestyle='--', label=f"Mean={df['sepal_length'].mean():.2f}")
ax.axvline(df['sepal_length'].median(), color='orange', linestyle='--', label=f"Median={df['sepal_length'].median():.2f}")
ax.set_title('Distribusi Sepal Length')
ax.legend()
plt.tight_layout()
plt.show()
```

**Interpretasi Langkah 3:** Mean dan median `sepal_length` berdekatan → distribusi mendekati simetris.


## Langkah 4: Boxplot & Violin Plot per Spesies

Bandingkan `petal_length` antar spesies dengan boxplot dan violin plot.


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, x='species', y='petal_length', palette='Set2', hue='species', legend=False, ax=axes[0])
axes[0].set_title('Boxplot Petal Length')
sns.violinplot(data=df, x='species', y='petal_length', palette='Set2', inner='box', hue='species', legend=False, ax=axes[1])
axes[1].set_title('Violin Plot Petal Length')
plt.tight_layout()
plt.show()
```

**Interpretasi Langkah 4:** *Setosa* punya `petal_length` jauh lebih kecil dan terpisah jelas dari spesies lain.


## Langkah 5: Matriks Korelasi Pearson

Hitung korelasi antar fitur numerik dan identifikasi pasangan dengan korelasi tertinggi.


```python
corr = df.drop('species', axis=1).corr(method='pearson')
print(corr.round(3))
mask = np.triu(np.ones(corr.shape, dtype=bool))
max_pair = corr.where(~mask).stack().idxmax()
print(f'\nKorelasi tertinggi: {max_pair} = {corr.loc[max_pair]:.3f}')
```

**Interpretasi Langkah 5:** **petal_width ↔ petal_length** korelasi tertinggi (0.963).


## Langkah 6: Scatter Plot & Heatmap

Scatter `petal_length` vs `petal_width` + heatmap korelasi.


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set2', ax=axes[0])
sns.regplot(data=df, x='petal_length', y='petal_width', scatter=False, color='gray', ax=axes[0])
axes[0].set_title('Scatter: Petal Length vs Petal Width')
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', vmin=-1, vmax=1, square=True, ax=axes[1])
axes[1].set_title('Heatmap Korelasi')
plt.tight_layout()
plt.show()
```

**Interpretasi Langkah 6:** Tren positif jelas pada scatter; setosa membentuk klaster terpisah.
