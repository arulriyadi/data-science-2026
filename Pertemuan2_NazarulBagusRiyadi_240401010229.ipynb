# Praktikum Pertemuan 2: Struktur Data Python, NumPy & Pandas


```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Q1 - jumlah baris, kolom, dan nama kolom
print(f"Jumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
print("Nama kolom:")
for col in df.columns:
    print(f"- {col}")
```

```python
# Q2 - missing value dan proporsi
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().mean()) * 100
missing_data = pd.DataFrame({
    'Jumlah Missing Values': missing_values,
    'Proporsi Missing Values (%)': missing_percentage.round(2)
})
missing_data = missing_data[missing_data['Jumlah Missing Values'] > 0]
print(missing_data)
```

```python
# Q3 - persen penumpang yang selamat
survived_count = df['Survived'].sum()
total_count = df.shape[0]
survived_percentage = (survived_count / total_count) * 100
print(f"Persentase penumpang yang selamat: {survived_percentage:.2f}%")
```

```python
# Q4 - penumpang wanita kelas 1
female_class_1 = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]
print(f"Jumlah: {len(female_class_1)}, Rata-rata usia: {female_class_1['Age'].mean():.2f}")
```

```python
# Q5 - tingkat keselamatan per kelas
survival_rate = df.groupby('Pclass')['Survived'].mean()
print(f"Tingkat keselamatan per kelas:\n{survival_rate}")
print(f"Kelas dengan tingkat keselamatan tertinggi: Kelas {survival_rate.idxmax()}")
```

```python
# Q6 - analisis tambahan jenis kelamin
survival_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_by_gender)
```
